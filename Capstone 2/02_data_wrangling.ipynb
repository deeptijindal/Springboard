### 2.6.1 Remove useless columns

### 2.6.2 Check for null values

Count the number of missing values in each column and sort them.

```python
#Call `pd.concat` to present these in a single table (DataFrame)
#Count (using `.sum()`) the number of missing values (`.isnull()`) in each column of 
#house_data as well as the percentages (using `.mean()` instead of `.sum()`).
missing = pd.concat([house_data.isnull().sum(), 100 * house_data.isnull().mean()], axis=1)
#Change column names to 'count' and '%'
missing.columns=['count', '%']
#Order them (increasing or decreasing) using sort_values
missing.sort_values(by=['count', '%'], ascending=False)
```
